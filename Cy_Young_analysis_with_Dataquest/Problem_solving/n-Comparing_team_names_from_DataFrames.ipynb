```python
import pandas as pd
```

```python
AL_cy_young_voting_DataFrame = pd.read_csv('../Data/AL_cy_young_voting.csv')
NL_cy_young_voting_DataFrame = pd.read_csv('../Data/NL_cy_young_voting.csv')

AL_pitching_stats_DataFrame = pd.read_csv('../Data/AL_pitching_stats.csv')
NL_pitching_stats_DataFrame = pd.read_csv('../Data/NL_pitching_stats.csv')

AL_standings_DataFrame = pd.read_csv('../Data/AL_standings.csv')
NL_standings_DataFrame = pd.read_csv('../Data/NL_standings.csv')
```

```python
AL_pitching_stats_DataFrame = AL_pitching_stats_DataFrame[AL_pitching_stats_DataFrame['Tm'].notna()]
NL_pitching_stats_DataFrame = NL_pitching_stats_DataFrame[NL_pitching_stats_DataFrame['Tm'].notna()]
```

```python
print(AL_cy_young_voting_DataFrame['Tm'].nunique(), end=' ')
print(AL_pitching_stats_DataFrame['Tm'].nunique(), end=' ')
print(AL_standings_DataFrame['Tm'].nunique())
```

```python
print(NL_cy_young_voting_DataFrame['Tm'].nunique(), end=' ')
print(NL_pitching_stats_DataFrame['Tm'].nunique(), end=' ')
print(NL_standings_DataFrame['Tm'].nunique())
```

```python
AL_c_y_v_names = list(AL_cy_young_voting_DataFrame['Tm'].unique())
AL_c_y_v_names.sort()
AL_c_y_v_names.extend(['--','--'])


AL_p_s_names = list(AL_pitching_stats_DataFrame['Tm'].unique())
AL_p_s_names.sort()
AL_p_s_names.append('--')

AL_s_names = list(AL_standings_DataFrame['Tm'].unique())
AL_s_names.sort()
```

```python
AL_team_names_DataFrame = pd.DataFrame({'c_y_v':AL_c_y_v_names,
                                        'p_s':AL_p_s_names,
                                        's':AL_s_names})
```

```python
NL_c_y_v_names = list(NL_cy_young_voting_DataFrame['Tm'].unique())
NL_c_y_v_names.sort()
NL_c_y_v_names.append('--')

NL_p_s_names = list(NL_pitching_stats_DataFrame['Tm'].unique())
NL_p_s_names.sort()

NL_s_names = list(NL_standings_DataFrame['Tm'].unique())
NL_s_names.sort()
NL_s_names.append('--')
```

```python
NL_team_names_DataFrame = pd.DataFrame({'c_y_v':NL_c_y_v_names,
                                        'p_s':NL_p_s_names,
                                        's':NL_s_names})
```
