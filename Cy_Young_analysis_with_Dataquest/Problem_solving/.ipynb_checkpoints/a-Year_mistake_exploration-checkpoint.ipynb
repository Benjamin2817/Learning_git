```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
seasons = list(range(2003,2023))
```

```python
for season in seasons:
    with open('../Data/Cy_Young_award_winners_webpages/Cy_Young_award_winner_{}.html'.format(season), encoding='utf-8') as file:
        page = file.read()
    
    current_page_soup = BeautifulSoup(page, 'html.parser')
    al_voting_table = current_page_soup.find(id='AL_CYA_voting')
    nl_voting_table = current_page_soup.find(id='NL_CYA_voting')
    
    al_voting_DataFrame = pd.read_html(str(al_voting_table))[0]
    nl_voting_DataFrame = pd.read_html(str(nl_voting_table))[0]
    
    print(season)
    print(al_voting_DataFrame.iloc[0][1])
    print(nl_voting_DataFrame.iloc[0][1])
    print('\n')
```
