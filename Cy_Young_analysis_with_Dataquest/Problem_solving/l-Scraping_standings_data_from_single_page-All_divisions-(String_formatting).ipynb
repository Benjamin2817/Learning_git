```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.baseball-reference.com/leagues/NL/2019-standings.shtml'
standings_page_response = requests.get(url)
```

```python
soup = BeautifulSoup(standings_page_response.text, 'html.parser')
```

```python
all_division_standings_for_season = []

for division in 'ECW':
    division_standings_code = soup.find(id='all_standings_{}'.format(division))
    division_standings_code_DataFrame = pd.read_html(str(division_standings_code))[0]
    all_division_standings_for_season.append(division_standings_code_DataFrame)
```

```python
all_division_standings_for_season_DataFrame = pd.concat(all_division_standings_for_season)
```

```python
all_division_standings_for_season_DataFrame['Season'] = 2019
```
